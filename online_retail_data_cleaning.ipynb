```python
# import necessary packages
import numpy as np
import pandas as pd
```

```python
# load data
data_df = pd.read_csv("Online_Retail.csv")
```

```python
# display the data
data_df.describe(include="all")
```

```python
# check the types of fields
data_df.dtypes
```

```python
# shape of the data
data_df.shape
```

```python
# check null values fields
data_df.isnull().sum()
```

```python
# delete samples with null values 
data_df_cleaned = data_df.dropna(axis=0, subset=["UserID"])
```

```python
# check null values after cleaning 
data_df_cleaned.isnull().sum()
```

```python
# check shape after cleaning 
data_df_cleaned.shape
```

```python
# add the 'total consumption' into data
data_df_cleaned["Total Consumption"] = data_df_cleaned["Quan"] * data_df_cleaned["ItemPrice"]
```

```python
# view data after adding the coloum
data_df_cleaned.head(20)
```

```python
# get the total consumption mean
total_comsumption_mean = data_df_cleaned["Total Consumption"].mean()
print("The average total consumption: ", total_comsumption_mean)
```

```python
# get users in the datasets
users = data_df_cleaned["UserID"].unique()
```

```python
# get total consumption per user
total_consumptions = []
for user in users:
    total_consumptions.append([user, 
                               data_df_cleaned.loc[data_df_cleaned["UserID"] == user, "Total Consumption"].sum()])

df_tc_user = pd.DataFrame(total_consumptions, columns=["UserID", "Total Consumptions"])
```

```python
# display the users with total cosumptions
df_tc_user.head(20)
```

```python
# get the total consumption mean
total_comsumption_mean = df_tc_user["Total Consumptions"].mean()
print("The average total consumption: ", total_comsumption_mean)
```

```python
# categorize the users by total consumption
bins = [-np.inf, 0, 1000, 2000, np.inf]
labels = ["Black", "Low", "Medium", "High"]

df_tc_user["Consumer Type"] = pd.cut(df_tc_user["Total Consumptions"], bins=bins, labels=labels)
```
